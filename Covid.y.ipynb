```python
# Q12: Top 5 countries having greater no.of covid cases:-

top_5_countries=df.nlargest(5,'Total Cases')
print(top_5_countries[['Country','Total Cases']])
```

```python
# Q13: Ploting the graph:-

top_5_countries=df.nlargest(5,"Total Cases")
plt.figure(figsize=(10,6))
sns.barplot(x='Country',y='Total Cases',data=top_5_countries)
plt.title("Top 5 Countries having largest Total covid-19 Cases")
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.show()
```

```python
# Q14: Least 5 countries which have least no. of covid cases:-

top_5_countries=df.nsmallest(5,'Total Cases')
print(top_5_countries[['Country','Total Cases']])
```

```python
# Q15: Plotting the graph:-

top_5_countries=df.nsmallest(5,'Total Cases')
plt.figure(figsize=(10,6))
plt.bar(top_5_countries['Country'],top_5_countries['Total Cases'])
plt.title("Top 5 Countries having least Total covid-19 Cases")
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.show()
```

```python
# Short analytical summary of the above dataset:

# 1. Overview of the Dataset:

The COVID-19 dataset provides detailed information about
the pandemic across various regions,containing data on 
cases, deaths, and related attributes. The dataset consist 
of 225 rows and 10 columns, each representing different
indicators such as Country.Total Cases, Total Deaths,Total
Recovered, Active Cases, and more. It offers a tructured
overview of how the virus impacted different regions globally.

On examining the dataset, only one missing value was detected
in the “Other names” column. Since this missing entry was
minor and not critical for numerical analysis, it could 
either be left as is or replaced.
```

```python
# 2.Key Observations;

From the COVID-19 dataset, it is evident that the United
States (USA) reported the highest number of confirmed 
cases, totaling approximately 81,839,052 cases.The USA also
recorded the highest number of deaths, reaching about 
1,008,222 deaths.

 Based on available information, the average fatality rate
across all listed countries is approx.1.44%, meaning that out
of every 100 confirmed cases, around 1 to 2 resulted in death.

This rate suggests that while COVID-19 spread widely, the
 majority of infected individuals recovered.
```

```python
# 3. Insights from Visualizations:

The visualizations of the COVID-19 dataset reveal several clear
patterns and trends. A bar graph comparing Total Cases by country 
shows that a few nations—particularly U.S., India, and Brazil—
dominated global case counts with both high population and testing 
capacities. Similarly, the graph of Total Deaths shows a parallel trend,
suggesting that higher case numbers often led to higher fatalities.
```

```python
# 4. Conclusion:

The COVID-19 dataset provides a clear understanding of how the pandemic
evolved and impacted different countries worldwide. The data indicates 
that countries like U.S., India, and Brazil experienced the highest 
number of confirmed cases and deaths, reflecting both their large 
populations and extensive testing capacities.The average fatality rate 
across all countries was found to be around 1.44%, suggesting that while 
the virus spread rapidly, the majority of infected individuals 
ultimately recovered.
```
